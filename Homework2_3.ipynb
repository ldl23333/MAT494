```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import pearsonr

# Gaussian distributed
np.random.seed(1)
n = 500
x = np.random.randn(n)
y = np.random.randn(n)**2

# joint distribution f(x,y) with marginal distributions on axes
sns.set()
sns.jointplot(x,y, kind = 'scatter', marginal_ticks=True)
plt.show()
```
